```python
import matplotlib.pyplot as plt
import numpy as np

fs = 1000
N = 1e5
amp = 2 * np.sqrt(2)
time = np.arange(N) / fs
data = amp * np.sin(2 * np.pi * 100 * time)

plt.specgram(data, NFFT=1024, Fs=fs, noverlap=512, cmap='plasma')
plt.title("Spectrogram")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), cmap='viridis')
plt.title("Streamplot")
plt.show()
```
